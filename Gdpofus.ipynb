Gross domestic product (GDP) is a measure of the market value of all the final goods and services produced in a period. GDP is an indicator of how well the economy is doing.
```python
! conda install bokeh

import pandas as pd

from bokeh.io import push_notebook, output_file,show, output_notebook

from bokeh.layouts import row

from bokeh.plotting import figure

output_notebook()
```

Defining the function make_dashboard 
You should care about the inputs to the dashboard 

```python
def make_dashboard(x, gdp_change, unemployment, title, file_name):
    output_file(file_name)
    output_notebook()
    p = figure(title=title, x_axis_label='year', y_axis_label='%')
    p.line(x.squeeze(), gdp_change.squeeze(), color="firebrick", line_width=4, legend="% GDP change")
    p.line(x.squeeze(), unemployment.squeeze(), line_width=4, legend="% unemployed")
    show(p)
```
